<h1 style="color:yellow">Matplotlib : <br>
pyplot.plot format strings and other options</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
# plt.plot(x,y,color='green',linewidth=5, linestyle='dashdot')
plt.plot(x,y,color='maroon',linewidth=3, linestyle='dotted')
```

```python
plt.plot(x,y,'rD--') #the third argument is like: 'color markeronpoints thingconnectingthem'
#order os this argument doesn't matter
```

**https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html**

```python
# Same effect as 'rD--' format string
plt.plot(x,y,color='red',marker='D',linestyle='--')
```

<h3 style="color:green">markersize</h3>

```python
plt.plot(x,y,color='red',marker='D',linestyle='--',markersize=40)
```

<h3 style="color:green">alpha property to control transparency of line chart</h3>

```python
plt.plot(x,y,color='red',marker='D',linestyle='--',alpha=0.4) # alpha can be specified on a scale 0 to 1
```
